# Activation Function

In order to let our neural network model complex patterns in the data, we have to add a non-linear property to the model. It often is an activation function, such as a *sigmoid* or a *tanh*, but these activation functions are bounded and their gradient is very low on the edges. Because we are going to manipulate high scalar values, we have to use an unbounded activation function, such as *ReLU*, $Φ(x) = max(0,x)$. But the issue with *ReLU* is that all the negative values become zero immediately, which decreases the ability of our model to train from the data. Hence the implementation of a leaky rectifier, *LReLU*:

$$Φ(x) = max(αx,x), \text{ with } 0 < α < 1.$$

We show below a few classic activation functions and the LReLU function mentionned before.

```python
import math
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
sb.set()
```

```python
def plot_fct(fct):
    """
    Plot a function in right range
    """
    x = np.arange(-5, 5, 0.2)
    y = fct(x)

    fig, ax = plt.subplots()
    ax.plot(x, y)
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.axhline(y=0, color='k', alpha=.4)
    ax.axvline(x=0, color='k', alpha=.4)
    plt.savefig(f"./pics/activation_fct_{fct.__name__}.png")
    plt.show()
```

## Sigmoid

```python
def sigmoid(x):
    return [1/(1+math.exp(-item)) for item in x]

plot_fct(sigmoid)
```

## Hyperbolic tangent

```python
def tangh(x):
    return [math.tanh(item) for item in x]

plot_fct(tangh)
```

## ReLU

```python
def ReLU(x):
    return [max(0, item) for item in x]

plot_fct(ReLU)
```

## LReLU

By implementing a Leaky Rectifier, we are able to take into account the negative valued pixels.

```python
alpha = .2
def LReLU(x):
    return [max(alpha * item, item) for item in x]

plot_fct(LReLU)
```
